```python
import cv2
import numpy as np
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt


def preprocess_image(img_path, size=(256, 256)):

    img = cv2.imread(img_path)


    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


    denoised = cv2.GaussianBlur(img, (5, 5), 0)


    resized = cv2.resize(denoised, size)

    return img, resized



def extract_hog(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    features, hog_image = hog(
        gray,
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        visualize=True,
        block_norm='L2-Hys'
    )


    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

    return features, hog_image_rescaled



def feature_matching(img1, img2):
    gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
    gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

    orb = cv2.ORB_create(500)

    kp1, des1 = orb.detectAndCompute(gray1, None)
    kp2, des2 = orb.detectAndCompute(gray2, None)

    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

    matches = bf.match(des1, des2)
    matches = sorted(matches, key=lambda x: x.distance)

    matched_img = cv2.drawMatches(
        img1, kp1, img2, kp2, matches[:30], None,
        flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS
    )

    return matched_img


if __name__ == "__main__":
    img_path1 = "img1.jpg"
    img_path2 = "img2.jpg"


    original1, img1 = preprocess_image(img_path1)
    original2, img2 = preprocess_image(img_path2)


    features, hog_img = extract_hog(img1)

    print("HOG feature vector length:", len(features))


    matched = feature_matching(img1, img2)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 3, 1)
    plt.title("Preprocessed Image")
    plt.imshow(img1)
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("HOG Features")
    plt.imshow(hog_img, cmap="gray")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("Feature Matching")
    plt.imshow(matched)
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
from ultralytics import YOLO


def train_model():
    # Load pretrained YOLOv8 nano model
    model = YOLO("yolov8n.pt")

    # Train on small COCO dataset (auto-downloads if not present)
    results = model.train(
        data="coco128.yaml",   # lightweight dataset
        epochs=10,             # keep small for testing
        imgsz=640,
        batch=16,
        device=0,              # use GPU (set 'cpu' if no GPU)
        project="runs/train",
        name="yolo_coco128",
        pretrained=True
    )

    return model


def evaluate_model(model):
    # Run validation
    metrics = model.val()

    print("\n===== EVALUATION METRICS =====")
    print(f"Precision: {metrics.box.mp:.4f}")
    print(f"Recall:    {metrics.box.mr:.4f}")
    print(f"mAP@0.5:   {metrics.box.map50:.4f}")
    print(f"mAP@0.5:0.95: {metrics.box.map:.4f}")

    return metrics


def test_on_image(model, image_path):
    # Run inference
    results = model.predict(
        source=image_path,
        conf=0.25,
        save=True,
        show=True
    )

    print("\nPrediction completed. Results saved in 'runs/detect/' folder.")
    return results


if __name__ == "__main__":
    # Step 1: Train
    model = train_model()

    # Step 2: Evaluate
    evaluate_model(model)

    # Step 3: Test on image
    test_image_path = "test.jpg"  # Replace with your image
    test_on_image(model, test_image_path)
```

```python
from IPython.display import Image, display
import os

def test_on_image(model, image_path):
    # Run inference and SAVE results image
    results = model.predict(
        source=image_path,
        conf=0.25,
        save=True,
        show=False
    )

    print("\nPrediction completed.")

    # YOLO saves output image automatically in this folder:
    save_dir = results[0].save_dir
    print(f"Saved in: {save_dir}")

    # Find saved image path
    files = os.listdir(save_dir)
    for f in files:
        if f.endswith(".jpg") or f.endswith(".png"):
            result_img_path = os.path.join(save_dir, f)
            break

    # Display image in notebook
    display(Image(filename=result_img_path))

    return results
```
