```python
from ultralytics import YOLO
import os

# Step 1: Load a pretrained YOLOv8 model
model = YOLO("yolov8n.pt")  # nano model (fast & lightweight)

# Step 2: Download a small aerial dataset (VisDrone - YOLO format)
dataset_url = "https://github.com/ultralytics/yolov5/releases/download/v1.0/VisDrone.yaml"

# Save dataset config file
os.system(f"wget {dataset_url}")

# Step 3: Train the model for 10 epochs
model.train(
    data="VisDrone.yaml",  # dataset config
    epochs=10,
    imgsz=640,
    batch=8,
    name="yolo_aerial_detection"
)

# Step 4: Run inference on sample images
results = model.predict(
    source="https://ultralytics.com/images/bus.jpg",  # replace with aerial image later
    save=True,
    conf=0.25
)

# Step 5: Print results
for r in results:
    print(r.boxes)
```
